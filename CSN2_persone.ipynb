```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', size = 14, family ='Sans-Serif')
key="1esZt19LIS654GhDG2cxAlC5L4qEZ0Yd0-jcA_VMlUA4"
url_csv_file = "https://docs.google.com/spreadsheet/ccc?key="+key+"&output=csv"
df = pd.read_csv(url_csv_file)
# https://docs.google.com/spreadsheets/d/1esZt19LIS654GhDG2cxAlC5L4qEZ0Yd0-jcA_VMlUA4/edit?usp=sharing
```

```python
fig, ax = plt.subplots(figsize=(20,6))
x = np.linspace(1, len(df),len(df))


ax.bar(x, df['RICERCATORI (FTE/PERS)'].values, color='blue', label='Ricercatori', alpha=0.5)
ax.bar(x, df['TECNOLOGI (FTE/PER)'].values, color='orange', label='Tecnologi', alpha=0.9)
ax.stem(x, df['TOT. PERS.'].values, 'k^', label='Persone', use_line_collection=True)
ax.plot(x, df['FTE'].values, 'g-', label='FTE')
ax.set_ylim(0,120)
ax.legend(loc=2)
ax = ax.twinx()
ax.plot(x, df['FTE / PERS.'].values*100, 'r--',label='FTE/Persone')
ax.set_ylim(0,100)
ax.legend(loc=1)
ax.set_xticks(x)
ax.set_xticklabels(df.SEZIONE.values)

plt.show()
```
